```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

We are going to create a function that provides us all the games in each region for the 2021/2022 season of Championnat National 3. 

```python
def get_matches_poules(poule): # User provides the name of the groupe with a letter corresponding to a region.
    
    h_teams = []
    a_teams = []
    scores = []
    no_journee = []
    for i in range(1,27): # Teams play 26 games in a season
        r = requests.get(f'https://www.bienpublic.com/sport/calendrier-resultats/football/national-national-3-groupe-{poule}/{i}')
        soup = BeautifulSoup(r.text, 'html.parser')

        
        for j in range(len(soup.select('td.scoreData'))):
            home_team = soup.select('td.firstTeam a')[j].get_text() # Home Team
            away_team = soup.select('td.secondTeam a')[j].get_text() # Away Team
            score = soup.select('td.scoreData')[j].get_text().strip() # Score
            
            h_teams.append(home_team)
            a_teams.append(away_team)
            scores.append(score)
            no_journee.append(i)
            

    
    df = pd.DataFrame(data = {'Equipe_Dom': h_teams,
                             'Equipe_Ext': a_teams,
                             'Score': scores,
                             'Journee': no_journee,
                              'Region': poule})
    
    df.to_csv(f'poule_{poule}.csv') # Export a dataframe of the region with all the games played during the season.
            
            
            
```
